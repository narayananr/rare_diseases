## Part 2: Which Genes Are Most Important?

Some genes are linked to many different diseases. A gene linked to many diseases might be  
involved in a fundamental biological process — when it breaks, many things go wrong.

```python
# Count how many diseases each gene is linked to
gene_counts = diseases['gene'].value_counts()

print("Top 10 genes linked to the most diseases:")
gene_counts.head(10)
```

```python
# Visualize the top 10 genes as a horizontal bar chart
top_genes = gene_counts.head(10)

plt.figure(figsize=(10, 6))
plt.barh(top_genes.index, top_genes.values, color='steelblue')
plt.xlabel('Number of Diseases')
plt.ylabel('Gene')
plt.title('Top 10 Genes Linked to Most Rare Diseases')
plt.gca().invert_yaxis()
plt.tight_layout()
plt.show()
```
