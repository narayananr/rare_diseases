## Part 2: Which Genes Are Most Important?

Some genes are linked to many different diseases. A gene linked to many diseases might be  
involved in a fundamental biological process — when it breaks, many things go wrong.

````python
# Count how many diseases each gene is linked to
gene_counts = diseases['gene'].value_counts()

print("Top 10 genes linked to the most diseases:")
gene_counts.head(10)
````

### Understanding the Chart Code

Before running the chart, let's break down what each line does:

```python
plt.figure(figsize=(10, 6))       # Create a canvas: 10 inches wide, 6 inches tall
plt.barh(top_genes.index,         # Horizontal bars — gene names on the Y axis
```

**Color options to try:** `'steelblue'`, `'coral'`, `'mediumseagreen'`, `'orchid'`, `'tomato'`, `'#4A90D9'`

````python
# Basic bar chart — run this first to see the default version
top_genes = gene_counts.head(10)

plt.figure(figsize=(10, 6))
plt.barh(top_genes.index, top_genes.values, color='steelblue')
plt.xlabel('Number of Diseases')
plt.ylabel('Gene')
plt.title('Top 10 Genes Linked to Most Rare Diseases')
plt.gca().invert_yaxis()
plt.tight_layout()
plt.show()
````

### Making a Better Chart

A good chart tells a clear story at a glance. Here are three improvements:

1. **Add value labels** — show the exact count on each bar so readers don't have to estimate
2. **Add a reference line** — highlight the average to give context
3. **Use a descriptive subtitle** — help the reader understand what they're looking at

Run the improved version below and compare it to the basic one above.

````python
# Improved chart with value labels, average line, and subtitle
top_genes = gene_counts.head(10)

fig, ax = plt.subplots(figsize=(10, 6))

bars = ax.barh(top_genes.index, top_genes.values, color='steelblue', edgecolor='white')
ax.invert_yaxis()

# Add value labels at the end of each bar
for bar in bars:
    width = bar.get_width()
    ax.text(width + 0.3,           # x position (just past the bar)
            bar.get_y() + bar.get_height() / 2,  # y position (middle of bar)
            f'{int(width)}',       # the label text
            va='center', fontsize=10)

# Add a vertical line at the average
avg = top_genes.values.mean()
ax.axvline(avg, color='red', linestyle='--', linewidth=1.2, label=f'Average: {avg:.1f}')
ax.legend()

# Titles and labels
ax.set_xlabel('Number of Rare Diseases Linked', fontsize=12)
ax.set_ylabel('Gene Symbol', fontsize=12)
ax.set_title('Top 10 Genes Linked to Most Rare Diseases', fontsize=14, fontweight='bold')
ax.set_xlim(0, top_genes.values.max() + 4)   # make room for labels

plt.tight_layout()
plt.show()

print(f"Average diseases per gene (top 10): {avg:.1f}")
````

### Customization: Color by Rank

Instead of one flat color, you can use a **color gradient** to visually encode rank — 
darker = more diseases. This makes the ranking easier to read at a glance.

````python
# Color gradient: darker bars = more diseases
top_genes = gene_counts.head(10)

# Generate shades from dark to light blue
import numpy as np
n = len(top_genes)
colors = [plt.cm.Blues(0.9 - i * 0.06) for i in range(n)]

fig, ax = plt.subplots(figsize=(10, 6))
bars = ax.barh(top_genes.index, top_genes.values, color=colors, edgecolor='white')
ax.invert_yaxis()

# Value labels
for bar in bars:
    width = bar.get_width()
    ax.text(width + 0.3, bar.get_y() + bar.get_height() / 2,
            f'{int(width)}', va='center', fontsize=10)

ax.set_xlabel('Number of Diseases', fontsize=12)
ax.set_ylabel('Gene Symbol', fontsize=12)
ax.set_title('Top 10 Genes — Darker = More Diseases', fontsize=14, fontweight='bold')
ax.set_xlim(0, top_genes.values.max() + 4)
plt.tight_layout()
plt.show()
````

### Chart Customization Cheat Sheet

| What to change | How |
|----------------|-----|
| Bar color | `color='coral'` — any color name or hex like `'#FF6B6B'` |
| Chart size | `figsize=(12, 8)` — wider/taller canvas |
| Font size | `fontsize=14` in any `ax.set_title()` or `ax.text()` call |
| Bar width (vertical chart) | `width=0.6` in `plt.bar()` |
| Remove top/right borders | `ax.spines['top'].set_visible(False)` |
| Grid lines | `ax.xaxis.grid(True, linestyle='--', alpha=0.5)` |
| Rotate x labels | `plt.xticks(rotation=45, ha='right')` |
