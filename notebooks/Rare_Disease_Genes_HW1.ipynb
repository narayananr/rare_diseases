## Part 2: Which Genes Are Most Important?

Some genes are linked to many different diseases. A gene linked to many diseases might be  
involved in a fundamental biological process — when it breaks, many things go wrong.

```python
# Count how many diseases each gene is linked to
gene_counts = diseases['gene'].value_counts()

print("Top 10 genes linked to the most diseases:")
gene_counts.head(10)
```

---
## Part 2: Building a Chart Step by Step

We'll build a chart one piece at a time. Run each cell and watch it improve.

### Step 1: A Plain Bar Chart

The simplest possible chart — just pass in the data. No labels, no title, no styling.

```python
top_genes = gene_counts.head(10)

plt.bar(top_genes.index, top_genes.values)
plt.show()
```

### Step 2: Add Axis Labels

Without labels a chart is unreadable. `xlabel` and `ylabel` tell the reader what each axis means.

```python
top_genes = gene_counts.head(10)

plt.bar(top_genes.index, top_genes.values)
plt.xlabel('Gene')
plt.ylabel('Number of Diseases')
plt.show()
```

### Step 3: Add a Title

The title answers "what am I looking at?" — make it specific.

```python
top_genes = gene_counts.head(10)

plt.bar(top_genes.index, top_genes.values)
plt.xlabel('Gene')
plt.ylabel('Number of Diseases')
plt.title('Top 10 Genes Linked to Most Rare Diseases')
plt.show()
```

### Step 4: Make It Horizontal

Gene names are long — they overlap on a vertical chart. `barh` flips the axes so names are easy to read.  
Notice: `xlabel` and `ylabel` swap, and we add `invert_yaxis()` so the highest bar is at the top.

```python
top_genes = gene_counts.head(10)

plt.barh(top_genes.index, top_genes.values)
plt.xlabel('Number of Diseases')
plt.ylabel('Gene')
plt.title('Top 10 Genes Linked to Most Rare Diseases')
plt.gca().invert_yaxis()
plt.show()
```

### Step 5: Add Value Labels on Each Bar

Readers shouldn't have to guess the exact number — put the count right on the bar.  
`bar.get_width()` gives the length of each bar, and `ax.text()` places a label there.

```python
top_genes = gene_counts.head(10)

fig, ax = plt.subplots()
bars = ax.barh(top_genes.index, top_genes.values)
ax.invert_yaxis()
ax.set_xlabel('Number of Diseases')
ax.set_ylabel('Gene')
ax.set_title('Top 10 Genes Linked to Most Rare Diseases')

# Add a number label at the end of every bar
for bar in bars:
    width = bar.get_width()
    ax.text(width + 0.2,                          # x: just past the bar tip
            bar.get_y() + bar.get_height() / 2,   # y: middle of the bar
            str(int(width)),                       # the label text
            va='center')

plt.show()
```

### Step 6: Color and Size

Pick a color and make the figure bigger so everything has room to breathe.  
Try changing `color=` to `'coral'`, `'mediumseagreen'`, `'orchid'`, or any hex code like `'#E67E22'`.

```python
top_genes = gene_counts.head(10)

fig, ax = plt.subplots(figsize=(10, 6))
bars = ax.barh(top_genes.index, top_genes.values, color='steelblue')
ax.invert_yaxis()
ax.set_xlabel('Number of Diseases')
ax.set_ylabel('Gene')
ax.set_title('Top 10 Genes Linked to Most Rare Diseases')

for bar in bars:
    width = bar.get_width()
    ax.text(width + 0.2, bar.get_y() + bar.get_height() / 2,
            str(int(width)), va='center')

plt.tight_layout()
plt.show()
```

### Quick Reference

| Line | What it does |
|------|-------------|
| `plt.bar(x, y)` | Vertical bars |
| `plt.barh(x, y)` | Horizontal bars |
| `plt.xlabel('...')` | Label the x axis |
| `plt.ylabel('...')` | Label the y axis |
| `plt.title('...')` | Add a title |
| `plt.gca().invert_yaxis()` | Put the highest bar at the top |
| `figsize=(10, 6)` | Canvas size in inches |
| `color='steelblue'` | Bar color |
| `plt.tight_layout()` | Prevent labels from being cut off |
